```python
# Runtime initialization
import numpy as np
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

# Interpolation problem statement

Given a set of data consisting of points (pairs of dependent and independent values), we would like to find an estimation function (interpolant) which passes through all points in the set.

Of course, not all interpolants are created equal; for example, if the set of points is a time series for the position of a car, a smooth interpolant would better correspond to the physics of the motion of the car than a jagged one. The processes that generate the data can be used to determine the quality of the interpolant.

# Linear interpolation

To generate a linear interpolant, we assume that points on the interpolant lie on a line connecting the point in the data set that are closest to the left and right, or are in the interpolating function itself.

Example: Generate the linear interpolant for the set {(0,0), (1,1), (2,1)}

```python
x = [0,1,2]
y = [0,1,1]
f = interp1d(x, y)
plt.plot(x, y, 'b-')
plt.plot(x, y, 'ko')
plt.title('Linear Interpolant for {(0,0), (1,1), (2,1)}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

# Cubic spline Interpolation

Linear interpolation, while simple, gives a jagged interpolant. An alternative is cubic spline interpolation, which connects subsequent points with cubic polynomials. Cubic spline interpolation results in a visually smooth interpolant with C<sup>2</sup> continuity. However, because the endpoints of the interpolant don't have this C<sup>2</sup> constraint (since there's nothing to connect to), there are 2 additional constraints which have to be given, as cubic polynomials have 4 coefficients (i.e. 4 degrees of freedom).

Example: Generate a cubic spline interpolant for the set {(0,0), (1,1), (2,1)}

```python
x = [0, 1, 2]
y = [0, 1, 1]

# Using bc_type = 'natural' constrains the second derivative at the endpoints to zero.
# This essentially makes the interpolant "look" linear at the endpoints.
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ko')
plt.title('Cubic Spline Interpolant for {(0,0), (1,1), (2,1)}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
